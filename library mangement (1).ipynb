```python
try:
    # Establish a connection to the MySQL server with the new database
    db_config['database'] = 'library1'
    connection = mysql.connector.connect(**db_config)

    # Create a cursor to execute SQL queries
    cursor = connection.cursor()

    # Create the 'library' table
    create_table_query = """
    CREATE TABLE IF NOT EXISTS library (
        member Varchar(255),
        prn_no varchar(255) PRIMARY KEY,
        Id  varchar(255),
        firstname VARCHAR(255),
        middlename VARCHAR(255),
        lastname VARCHAR(255),
        address VARCHAR(255),
        mobilenumber VARCHAR(20),
        bookid varchar(255),
        title VARCHAR(255),
        author VARCHAR(255),
        price varchar(255),
        dateborrowed varchar(255),
        datedue varchar(255),
        latereturn VARCHAR(255),
        overdue varchar(255)
    )
    """

    cursor.execute(create_table_query)
    print("Table 'library' created successfully in the database.")

    # Commit the changes and close the connection
    connection.commit()
    cursor.close()
    connection.close()

except mysql.connector.Error as err:
    print(f"Error: {err}")
```
